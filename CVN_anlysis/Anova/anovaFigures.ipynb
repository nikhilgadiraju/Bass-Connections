```python
df = pd.read_csv("posthoc_standard.csv")
df[df["Regions"] == "DIEnt_R"]
df[df["Regions"] == "VIEnt_L"]
```

```python
df_vol = pd.read_csv("voxelVolumes_treatments.csv")
df_vol = df_vol.replace({"Treatment": "wheel_only"}, "voluntary")
df_vol = df_vol.replace({"Treatment": "treadmill"}, "voluntary\n+ enforced")
```

```python
df_vol.iloc[:, 4:] = df_vol.iloc[:, 4:] * 0.001
```

```python
sns.set_theme()
sns.set(font_scale = 1.2)
df = df.sort_values(by=['ST'], ascending = False)

fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey = False, sharex = True)
plt.grid()
fig.suptitle('Most Significant Regions via Post-Hoc Analysis\n(Sedentary vs. Voluntary + Enforced Exercise)', y=1.15, fontsize = 16)

name = "CEnt_R"
print(name)
sns.violinplot(ax = axes[0], x = "Treatment", y = name, data = df_vol.filter(['Treatment', name], axis=1)).set_ylabel("")
axes[0].set_title(" ".join(("Caudomedial Entorhinal Cortex (R)".split("_"))),fontsize = 14)
axes[0].set_xlabel(" ")

name = "Cn_L"
print(name)
sns.violinplot(ax = axes[1], x = "Treatment", y = name, data = df_vol.filter(['Treatment', name], axis=1)).set_ylabel("")
axes[1].set_title(" ".join(("Cuneiform Nucleus (L)".split("_"))),fontsize = 14)
axes[1].set_xlabel(" ")

name = "Au1_R"
print(name)
plt.grid()
sns.violinplot(ax = axes[2], x = "Treatment", y = name, data = df_vol.filter(['Treatment', name], axis=1)).set_ylabel("")
axes[2].set_title(" ".join(("Primary Auditory Cortex (R)".split("_"))),fontsize = 14)
axes[2].set_xlabel(" ")

plt.savefig('st.png')
```
