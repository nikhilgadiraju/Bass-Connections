```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ls
```

```python
df = pd.read_csv("posthoc_standard.csv")
```

```python
df = df.sort_values(by=['ST'], ascending = False)
```

```python
df_vol = pd.read_csv("voxelVolumes_treatments.csv")
df_vol
```

```python
subsec = df_vol.filter(['Treatment', 'DIEnt_R'], axis=1)
subsec_sed = subsec[subsec['Treatment'] == 'sedentary']
subsec_tread = subsec[subsec['Treatment'] == 'treadmill']
```

```python
fig, ax = plt.subplots()
ax.violinplot([subsec_sed.DIEnt_R, subsec_tread.DIEnt_R])
```

```python
sns.violinplot(x = "Treatment", y = "DIEnt_R", data = subsec)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey = False)
plt.grid()
fig.suptitle('Effect of Exercise on Voxel Count across Brain Regions\n(Sedentary vs. Treadmill)', y=1.08)

name = df.iloc[-1].Regions
sns.violinplot(ax = axes[0], x = "Treatment", y = name, data = df_vol.filter(['Treatment', name], axis=1))
axes[0].set_title(" ".join(("Dorsal_Intermediate_Entorhinal_Cortex (R)".split("_"))))

name = df.iloc[-2].Regions
sns.violinplot(ax = axes[1], x = "Treatment", y = name, data = df_vol.filter(['Treatment', name], axis=1))
axes[1].set_title(" ".join(("Ventral_Intermediate_Entorhinal_Cortex (R)".split("_"))))
```
